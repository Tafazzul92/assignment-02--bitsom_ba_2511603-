```python
sentences = [
    'The batsman scored a brilliant century in the final match.',
    'The cricket team practiced hard before the world cup.',
    'The fast bowler delivered an impressive spell in the first innings.',
    'The stadium roared as the captain hit the winning shot.',
    'I learned a new recipe for making creamy pasta.',
    'Fresh herbs can completely transform the flavor of a dish.',
    'The chef slow-cooked the meat for six hours to make it tender.',
    'Strong passwords help protect accounts from hacking attempts.',
    'The company implemented multi-factor authentication for better security.',
    'Cybersecurity analysts monitor networks for suspicious activity.'
]
len(sentences)
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings = model.encode(sentences, convert_to_tensor=True)
embeddings.shape
```

```python
similarity_matrix = util.cos_sim(embeddings, embeddings).cpu().numpy()
similarity_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=False, cmap='viridis')
plt.title('Cosine Similarity Heatmap (10x10)')
plt.xlabel('Sentence Index')
plt.ylabel('Sentence Index')
plt.show()
```

```python
query = 'The bowler took three wickets in one over'
query_emb = model.encode(query, convert_to_tensor=True)
scores = util.cos_sim(query_emb, embeddings)[0]
top_results = scores.topk(2)
print('Query:', query)
print('
Top 2 Most Similar Sentences:
')
for idx, score in zip(top_results.indices, top_results.values):
    print(f'Sentence: {sentences[idx]}')
    print(f'Similarity Score: {float(score):.4f}
')
```
